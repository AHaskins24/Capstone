Consumer behavior data from jpmorgan

```python
import pandas as pd
import matplotlib.pyplot as plt

#read in the data
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\consumerbehavior.csv")
data.head()
```

```python
categories = data.columns[1:]
values = data.iloc[0, 1:].tolist()

colors = ['#ba9990', '#77a5a2', '#336d7c', '#083c4e', '#062e3b']

fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.6), colors=colors)

for autotext in autotexts:
    autotext.set_color('#f1e7e4')
    autotext.set_fontweight('bold')

centre_circle = plt.Circle((0, 0), 0.01, color='white')
ax.add_artist(centre_circle)
ax.set_aspect('equal')

ax.set_title('Consumer Behavior during Covid')

plt.show()

max_values = {}
for column in categories:
    max_value = data[column].max()
    max_date = data.loc[data[column] == max_value, 'Category'].values[0]
    max_values[column] = (max_value, max_date)
    print("Maximum value for", column + ":", max_value)
    print("Date for maximum value in", column + ":", max_date)
    print()

max_health_personal_care = max_values['Health & personal care'][0]
max_health_personal_care_date = max_values['Health & personal care'][1]
print("Maximum value in 'Health & personal care' column:", max_health_personal_care)
print("Date for maximum value in 'Health & personal care' column:", max_health_personal_care_date)

```

```python
#line plot
plt.figure(figsize=(25, 6))
for i, column in enumerate(categories):
    plt.plot(data['Category'], data[column], label=column, color=colors[i])

plt.title('US Market Growth across staple categories')
plt.legend()
plt.show()
```
