```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\covidcbjpmorgan.csv")
data.head(10)
```

```python
categories = data['Category']
ecommerce_growth = data['E-commerce as % of Sales']

colors = ['#52004e', '#5e004f', '#8f0053', '#d92654', '#dd5875', '#eec2b7', '#f1e7e4']

plt.figure(figsize=(12, 6))
plt.plot(categories, ecommerce_growth, color=colors[2], marker='o')
plt.title('E-commerce Growth over Time')
plt.ylabel('E-commerce as % of Sales')
plt.grid(True)
plt.xticks()
plt.show()
```
