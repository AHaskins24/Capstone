```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



# Read in data
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Percent_Change_in_Consumer_Spending__2020-2022.csv")
data.head(1)

```

```python
averages = data.iloc[:, 1:].mean()

column_with_max_average = averages.idxmax()
max_average = averages[column_with_max_average]

column_with_min_average = averages.idxmin()
min_average = averages[column_with_min_average]

print("Column with the highest average percent change:", column_with_max_average)
print("Average of", column_with_max_average + ":", round(max_average))
print("Column with the lowest average percent change:", column_with_min_average)
print("Average of", column_with_min_average + ":", round(min_average))
```

```python
#How does the overall consumer spending change over time during the COVID period?

#plot the overall consumer spending over time
data['Date'] = pd.to_datetime(data['Date'])
data.plot(x='Date', y='All merchant category codes spending', kind='line', color='#083c4e')
plt.xlabel('Date')
plt.ylabel('Consumer Spending')
plt.title('Overall Consumer Spending over Time')
plt.show()
```

```python
# Convert 'Date' 
data['Date'] = pd.to_datetime(data['Date'])

# Find the minimum and maximum dates
min_date = data['Date'].min()
max_date = data['Date'].max()

print("Minimum Date:", min_date)
print("Maximum Date:", max_date)
```

```python
# Filter the data based on the specified date range
filtered_data = data[(data['Date'] >= min_date) & (data['Date'] <= max_date)]
column_names = filtered_data.columns[3:4]

#line plot 
plt.figure(figsize=(12, 6))

color = '#8F0053'

for column in column_names:
    plt.plot(filtered_data['Date'], filtered_data[column], label=column, color=color)


plt.ylabel('Percentage Change')
plt.title('Changes in Accomodation and Food Service Spending over Time')

plt.show()
```

```python
import pandas as pd

# Define a dictionary mapping FIPS codes to state names
fips_to_state = {
    '900': 'Connecticut',
    '2500': 'Massachusetts',
    '3400': 'New Jersey',
    '3600': 'New York',
    '4400': 'Rhode Island'
}

# Define the file paths for the original and new CSV files
file_path_original = r'C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Percent_Change_in_Consumer_Spending__2020-2022.csv'
file_path_new = r'C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Percent_Change_in_Consumer_Spending2.csv'

# Read the original CSV file into a DataFrame
df_original = pd.read_csv(file_path_original)

# Read the new CSV file into a DataFrame
df_new = pd.read_csv(file_path_new)

# Perform the merge based on the common column (if applicable)
merged_df = pd.merge(df_original, df_new, on='State FIPS code')

# Translate the FIPS codes and populate the "State" column
merged_df['State'] = merged_df['State FIPS code'].map(fips_to_state)

# Save the merged DataFrame to a new CSV file
output_file_path = r'C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Merged_Consumer_Spending.csv'
merged_df.to_csv(output_file_path, index=False)

# Print the merged DataFrame
merged_df.head()

```

```python
fips_to_state = {
    '90': 'Connecticut',
    '25': 'Massachusetts',
    '34': 'New Jersey',
    '36': 'New York',
    '44': 'Rhode Island'
}

#new CSV file
file_path_new = r'C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Percent_Change_in_Consumer_Spending2.csv'

df_new = pd.read_csv(file_path_new)

#replace
df_new['State'] = df_new['State FIPS code'].astype(str).str[:2].map(fips_to_state)

# Save new DataFrame to a new CSV file
newfp = r'C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Percent Change in Consumer Spending, 2020-2022\Updated_Consumer_Spending2.csv'
df_new.to_csv(newfp, index=False)


df_new.head()
```

```python
column_names = df_new.columns.tolist()
print(column_names)
```

```python
# Find the state with the most growth in grocery and food store (GRF) spending over time
state_most_growth_GRF = df_new.groupby('State')['Grocery and food store (GRF)  spending'].sum().idxmax()
print("State with the Most Growth in Grocery and Food Store (GRF) Spending: ", state_most_growth_GRF)

# Find the state with the most growth in healthcare spending over time
state_most_growth_healthcare = df_new.groupby('State')['Health care and social assistance (HCS) spending '].sum().idxmax()
print("State with the Most Growth in Healthcare Spending: ", state_most_growth_healthcare)

# Convert 'Date' column to datetime
df_new['Date'] = pd.to_datetime(df_new['Date'])

```

```python
# Calculate the sum of spending per year for grocery and food store (GRF) spending and round to the nearest integer
sum_spending_per_year_GRF = df_new.groupby(['State', df_new['Date'].dt.year])['Grocery and food store (GRF)  spending'].sum().round(decimals=1)
print("Sum of Grocery and Food Store (GRF) Spending per Year (Rounded to Nearest Integer):")
print(sum_spending_per_year_GRF)

```

```python
# Define the colors
colors = ['#e2debc', '#dfb399', '#ba9990', '#adc4af', '#77a5a2', '#336d7c', '#083c4e', '#062e3b']

# Extract the year from the 'Date' column
df_new['Year'] = pd.to_datetime(df_new['Date'], format='%Y%m%d').dt.year

# Calculate the sum of spending per year for grocery and food store (GRF) spending and round to the nearest integer
sum_spending_per_year_GRF = df_new.groupby(['State', 'Year'])['Grocery and food store (GRF)  spending'].sum().round(decimals=0)

# Convert the grouped data to a DataFrame
df_sum_spending_GRF = sum_spending_per_year_GRF.reset_index()

# Plot the line graph
plt.figure(figsize=(12, 6))
for i, state in enumerate(df_sum_spending_GRF['State'].unique()):
    data = df_sum_spending_GRF[df_sum_spending_GRF['State'] == state]
    plt.plot(data['Year'], data['Grocery and food store (GRF)  spending'], color=colors[i % len(colors)], label=state)

plt.xlabel('Year')
plt.ylabel('Sum of Grocery and Food Store (GRF) Spending')
plt.title('Sum of Grocery and Food Store (GRF) Spending per Year')
plt.legend()
plt.show()


```
