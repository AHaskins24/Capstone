Real Personal Consumption Expenditures: Goods

        â€¢https://fred.stlouisfed.org/series/DGDSRX1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in the csv
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Dashboard data\DGDSRX1.csv")

data.head()
```

```python
# Convert 'DATE' column to datetime
data['DATE'] = pd.to_datetime(data['DATE'])



# Line Graph
plt.plot(data['DATE'], data['DGDSRX1'])
plt.ylabel('Billions of Chained 2012 Dollars')
plt.title('Growth of Goods over Time')
plt.show()
```

```python
# Convert 'DATE' column to datetime
data['DATE'] = pd.to_datetime(data['DATE'])

# Create scatter plot
plt.scatter(data['DATE'], data['DGDSRX1'])
plt.ylabel('Billions of Chained 2012 Dollars')
plt.title('Personal Consumption of Goods over Time')
plt.xticks()  
plt.tight_layout()  
plt.show()
```

Real personal consumption expenditures: Food

        â€¢https://fred.stlouisfed.org/series/DFXARX1M020SBEA

```python
pcefood = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\Dashboard data\pcefood.csv")
pcefood.head()
```

```python
pcefood['DATE'] = pd.to_datetime(pcefood['DATE'])

#maximum value and its index
max_value = pcefood['DFXARX1M020SBEA'].max()
max_index = pcefood['DFXARX1M020SBEA'].idxmax()
max_date = pcefood.loc[max_index, 'DATE']

#minimum value
min_value = pcefood['DFXARX1M020SBEA'].min()
min_index = pcefood['DFXARX1M020SBEA'].idxmin()
min_date = pcefood.loc[min_index, 'DATE']

print("Maximum amount of Food Expenditure:", max_value)
print("Date of the maximum Food Expenditure:", max_date)
print("Minimun amount spent on food:" , min_value)
print("Date of minimum amount of food purchase:" , min_date)
```

```python
pcefood['DATE'] = pd.to_datetime(pcefood['DATE'])

# Create a scatter plot with custom color
plt.figure(figsize=(12, 6))
plt.scatter(pcefood['DATE'], pcefood['DFXARX1M020SBEA'], color='#52004e')
plt.ylabel('PCE Food')
plt.title('PCE Food Growth Over Time')
plt.show()

```
