This is the Notebook to explore the Topic of World's Happiness. The dataset is from: https://www.kaggle.com/datasets/usamabuttar/world-happiness-report-2005-present

```python
#read in the data
import pandas as pd

# Read the CSV file
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\World Happiness Report.csv")

# Explore the data
data.head()
```

```python
# Filter the data for rows with "United States" in the "Country Name" column
us_data = data[data['Country Name'] == 'United States']

# Display the filtered data
us_data
```

```python
import matplotlib.pyplot as plt

# Define the custom color palette
colors = ["#000000", "#F1E7E4", "#15003D", "#52004E", "#8F0053", "#D92654", "#FDBFB0"]


```

```python
# Plotting a Line Plot
plt.plot(us_data['Year'], us_data['Generosity'])
plt.xlabel('Year')
plt.ylabel('Generosity')
plt.title('Generosity Over Time (United States)')
plt.show()


```

```python
# Define the custom color palette
colors = ["#15003D", "#52004E", "#8F0053", "#D92654", "#FDBFB0"]

# Plotting a Bar Plot with custom color palette
plt.bar(us_data['Year'], us_data['Generosity'], color=colors)
plt.xlabel('Year')
plt.ylabel('Generosity')
plt.title('Generosity Over Time (United States)')
plt.show()
```
