This is the Notebook to explore the Topic of World's Happiness. The dataset is from: https://www.kaggle.com/datasets/usamabuttar/world-happiness-report-2005-present

```python
#read in the data
import pandas as pd

#Import packages
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\World Happiness Report.csv")

# Explore the data
data.head()
```

```python
# Read the CSV file
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\World Happiness Report.csv")

# Calculate the average generosity by grouping the data by "Country Name"
avg_generosity = data.groupby('Country Name')['Generosity'].mean().reset_index()

# Plotting a Bar Plot of the average generosity
plt.figure(figsize=(10, 6))
plt.bar(avg_generosity['Country Name'], avg_generosity['Generosity'])
plt.xlabel('Country')
plt.ylabel('Average Generosity')
plt.title('Average Generosity by Country')
plt.xticks(rotation=90, fontsize=8)  # Set fontsize to adjust the x-axis labels
plt.tight_layout()
plt.show()

```

```python
# Group the data by "Country Name" and find the maximum and minimum generosity values
max_generosity = data.groupby('Country Name')['Generosity'].max().reset_index()
min_generosity = data.groupby('Country Name')['Generosity'].min().reset_index()

# Display the maximum and minimum generosity values per country
print("Maximum Generosity:")
print(max_generosity)
print("\nMinimum Generosity:")
print(min_generosity)
```

```python
# Group the data by "Country Name" and find the maximum and minimum generosity values
max_generosity = data.groupby('Country Name')['Generosity'].max().reset_index()
min_generosity = data.groupby('Country Name')['Generosity'].min().reset_index()

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the maximum and minimum generosity values
ax.bar(max_generosity['Country Name'], max_generosity['Generosity'], label='Maximum Generosity')
ax.bar(min_generosity['Country Name'], min_generosity['Generosity'], label='Minimum Generosity')

# Set labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Generosity')
ax.set_title('Maximum and Minimum Generosity by Country')
ax.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the average generosity by grouping the data by "Country Name"
avg_generosity = data.groupby('Country Name')['Generosity'].mean().reset_index()

# Find the country with the highest average generosity
highest_avg_generosity_row = avg_generosity.loc[avg_generosity['Generosity'].idxmax()]
highest_avg_generosity_country = highest_avg_generosity_row['Country Name']
highest_avg_generosity_value = highest_avg_generosity_row['Generosity']

# Print the country with the highest average generosity and its value
print("Country with the highest average generosity:", highest_avg_generosity_country)
print("Average generosity value:", highest_avg_generosity_value)

```

```python
# Calculate the average generosity by grouping the data by "Country Name"
avg_generosity = data.groupby('Country Name')['Generosity'].mean().reset_index()

# Find the country with the lowest average generosity
lowest_avg_generosity_row = avg_generosity.loc[avg_generosity['Generosity'].idxmin()]
lowest_avg_generosity_country = lowest_avg_generosity_row['Country Name']
lowest_avg_generosity_value = lowest_avg_generosity_row['Generosity']

# Print the country with the lowest average generosity and its value
print("Country with the lowest average generosity:", lowest_avg_generosity_country)
print("Average generosity value:", lowest_avg_generosity_value)
```
