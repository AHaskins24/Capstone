This is the Notebook to explore the Topic of World's Happiness. The dataset is from: https://www.kaggle.com/datasets/usamabuttar/world-happiness-report-2005-present

```python
#read in the data
import pandas as pd
import numpy as np

#Import packages
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\World Happiness Report.csv")

# Explore the data
data.head()
```

```python
# Read the CSV file
data = pd.read_csv(r"C:\Users\ahask\Documents\DA8\Capstone\Capstone\Data\World Happiness Report.csv")

# Calculate the average generosity by grouping the data by "Country Name"
avg_generosity = data.groupby('Country Name')['Generosity'].mean().reset_index()

# Plotting a Bar Plot of the average generosity
plt.figure(figsize=(8, 35))
plt.barh(avg_generosity['Country Name'], avg_generosity['Generosity'])
plt.xlabel('Country')
plt.ylabel('Average Generosity')
plt.title('Average Generosity by Country')
plt.xticks(fontsize=8)  # Set fontsize to adjust the x-axis labels
plt.tight_layout()
plt.show()

```

```python
# Find the country with the highest and lowest generosity values for each year
max_generosity = data.groupby('Year')['Generosity'].idxmax()
min_generosity = data.groupby('Year')['Generosity'].idxmin()

# Get the country names corresponding to the highest and lowest generosity values
country_max_generosity = data.loc[max_generosity, ['Year', 'Country Name', 'Generosity']]
country_min_generosity = data.loc[min_generosity, ['Year', 'Country Name', 'Generosity']]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the highest and lowest generosity values over time
ax.plot(country_max_generosity['Year'], country_max_generosity['Generosity'], label='Highest Generosity')
ax.plot(country_min_generosity['Year'], country_min_generosity['Generosity'], label='Lowest Generosity')

# Annotate country names for the highest generosity values
for x, y, country in zip(country_max_generosity['Year'], country_max_generosity['Generosity'], country_max_generosity['Country Name']):
    ax.annotate(country, xy=(x, y), xytext=(5, 0), textcoords='offset points', ha='left', va='center')

# Annotate country names for the lowest generosity values
for x, y, country in zip(country_min_generosity['Year'], country_min_generosity['Generosity'], country_min_generosity['Country Name']):
    ax.annotate(country, xy=(x, y), xytext=(-5, 0), textcoords='offset points', ha='right', va='center')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Generosity')
ax.set_title('Highest and Lowest Generosity by Year')
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()


```
